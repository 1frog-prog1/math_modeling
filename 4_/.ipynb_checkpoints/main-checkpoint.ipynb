```python
import numpy as np 
import matplotlib.pyplot as plt 
import random as rnd
```

```python
w = 10
```

```python
def right(t, x):
     return np.array([
        x[2], 
        x[3],
        2 * w * x[3],
        -2 * w * x[2]
     ])

def solve_odu2(f, y0, a, b, h):
    num = int(np.ceil((b - a) / h))
    t = np.linspace(a, b, num=num, endpoint=False)
    x = [y0] * num

    for i in range(num - 1):
        k0 = f(t[i], x[i])
        k1 = f(t[i] + h/2, x[i] + h/2*k0)
        k2 = f(t[i] + h/2, x[i] + h/2*k1)
        k3 = f(t[i] + h,x[i] + h*k2)
        x[i + 1] = x[i] + h/6 * (k0 + 2*k1 + 2*k2 + k3)

    return t, np.array(x)        
```

```python
t0 = 0 
tn = 10
n = 10000
h = (tn - t0) / n
```

```python
g = w / 100
fig, ax = plt.subplots(1, 2, figsize=(15, 9))
for i in range(4):
    x0 = [8, 4, rnd.randint(-3, 3), rnd.randint(-3, 3) ]
    t, x = solve_odu2(right, x0, t0, tn, h)
    ax[0].plot(x[:, 0], x[:, 1])
ax[0].plot(8, 4, marker='o', color='black')
# ax[0] = plt.gca()
ax[0].set_aspect('equal', adjustable='box')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('Траектории движения')

ax[1].set_aspect('equal', adjustable='box')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_title('Погрешность')
```
