```python
import numpy as np 
import matplotlib.pyplot as plt 
import random as rnd
```

```python
w = 10
```

```python
def right(t, x):
     return np.array([
        x[2], 
        x[3],
        2 * w * x[3],
        -2 * w * x[2]
     ])

def solve_odu2(f, y0, a, b, h):
    num = int(np.ceil((b - a) / h))
    t = np.linspace(a, b, num=num, endpoint=False)
    x = [y0] * num

    for i in range(num - 1):
        k0 = f(t[i], x[i])
        k1 = f(t[i] + h/2, x[i] + h/2*k0)
        k2 = f(t[i] + h/2, x[i] + h/2*k1)
        k3 = f(t[i] + h,x[i] + h*k2)
        x[i + 1] = x[i] + h/6 * (k0 + 2*k1 + 2*k2 + k3)

    return t, np.array(x)        
```

```python
t0 = 0 
tn = 10
n = 100000
h = (tn - t0) / n
```

```python
xy = [(-2, -2), (3, -2), (1, -5), (0, -6)]
```

```python
w = 1.5
fig, ax = plt.subplots(1, 2, figsize=(15, 9))
for i in range(4):
    x0 = [8, 4, xy[i][0], xy[i][1]]
    t, x = solve_odu2(right, x0, t0, tn, h)
    ax[0].plot(x[:, 0], x[:, 1], label=f'{(x0[2], x0[3])}')
    err =  (np.power(x[:, 2], 2) + np.power(x[:, 3], 2) - x0[2]**2 - x0[3]**2) / (x0[2]**2 + x0[3]**2)
    ax[1].plot(t, err, label=f'{(x0[2], x0[3])}', alpha=0.7)

ax[0].plot(8, 4, marker='o', color='black')
# ax[0] = plt.gca()
ax[0].set_aspect('equal', adjustable='box')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('Траектории движения')
ax[0].legend()
ax[0].set_xlim([2, 9])
ax[0].set_ylim([0, 7])

ax[1].legend()
ax[1].set_xlabel('t')
ax[1].set_ylabel('error')
ax[1].set_title('Погрешность')

fig.suptitle(f'Угловая скорость $w = {w}$', fontsize=18);
fig.savefig('/home/master/Data/math_modeling/my_labs/4_rotation/tex/pics/building_graphics/10.png')
```
