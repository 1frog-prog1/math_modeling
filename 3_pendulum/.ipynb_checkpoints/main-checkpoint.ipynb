```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def right0(t, uz):
    return np.array([
        uz[1],
        -w**2 * np.sin(uz[0])
    ])

def right1(t, uz):
    return np.array([
        uz[1],
        -w**2 * uz[0]
    ])

def right_friction(t, uz):
    return np.array([
        uz[1],
        -w**2 * uz[0] - k * uz[1]
    ])

def right_pendulums(t, uz):
    return np.array([
        uz[1],
        -w**2 * uz[0] + af * np.cos(wf * t)
    ])

def right_pendulums_frict(t, uz):
    return np.array([
        uz[1], 
        -w**2 * uz[0] - k * uz[1] + af * np.cos(wf * t)
    ])

def solve_odu2(f, t, u):
    a = [u0[0]]
    z = [u0[1]]
    u = u0
    for i in range(len(t) - 1):
        k0 = f(t[i], u)
        k1 = f(t[i] + h/2, u + h/2*k0)
        k2 = f(t[i] + h/2, u + h/2*k1)
        k3 = f(t[i] + h,u + h*k2)
        u = u + h/6 * (k0 + 2*k1 + 2*k2 + k3)

        a.append(u[0])
        z.append(u[1])
        
    return a, z    
```

```python
t0, tn = 0, 10
h = 0.01
g = 9.8
L = 2
w = np.sqrt(g / L)

t = np.arange(t0, tn + h/2, h)
```

```python
simple_path = '/home/master/Data/math_modeling/my_labs/3_pendulum/tex/pics/pendulum_examples/simple'
friction_path = '/home/master/Data/math_modeling/my_labs/3_pendulum/tex/pics/pendulum_examples/friction'
pendulums_path = '/home/master/Data/math_modeling/my_labs/3_pendulum/tex/pics/pendulum_examples/pendulums'
pendulums_fric_path = '/home/master/Data/math_modeling/my_labs/3_pendulum/tex/pics/pendulum_examples/pendulums_fric'
```

```python
u0 = np.array([np.pi / 3, 0])
a0, a0_div = solve_odu2(right0, t, u0)
a1, a1_div = solve_odu2(right1, t, u0)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].plot(t, a0, label=',  ')
axs[0].plot(t, a1, label=' ')
axs[0].set_title('  $a(t)$')
axs[0].set_xlabel('$t$')
axs[0].set_ylabel('$a$')
axs[0].set_ylim([-1.2, 1.5])
axs[0].legend()

axs[1].plot(a0, a0_div, label=',  ')
axs[1].plot(a1, a1_div, label=' ')
axs[1].set_title("  $a'(a)$")
axs[1].set_xlabel('$a$')
axs[1].set_ylabel("$a'$")
axs[1].legend()
fig.suptitle('  $ a_0 = \dfrac{\pi}{3} $')
fig.savefig(f'{simple_path}/pi_3.png');
```
