## Построение математической модели
Для того, чтоб проанализировать изменение температуры с течением времени, необходимо узнать, засчёт чего происходит повышение или понижение температуры, какие параметры влияют на скорость изменение температуры.

Для начала вспомним, что изменение температуры тела на $\Delta T$ вместе с массой тела $m$ и удельной теплоемкостью вещества тела $c$ связано с выделяемом телом количеством теплоты $\Delta Q$ по формуле:
$$ \Delta Q = cm \Delta T.$$

Изменение количества теплоты уже конкретно электрического нагревателя $ \Delta Q$ зависит от нескольких процессов: увеличение тепла путём преобразования электрической энергии в тепловую ($\Delta Q_н$), уменьшения засчёт теплопроводимости ($\Delta Q_т$) и засчёт теплового излучения ($ \Delta Q_{изл}$).
$$ \Delta Q = \Delta Q_н - \Delta Q_т - \Delta Q_{изл} $$

Количество тепла $Q_н$, выделенной путем преобразование электрической энергии в тепловую за промежуток времени $\Delta t$, связано с мощностью сети $P$ в сети по формуле:
$$ \Delta Q_н = P \Delta t. $$

Излучение -- это

Тепло, которое отдаётся внешней среде в резултате каждого из процессов считается по формулам: 
$$ \Delta Q_т =  $$
$$ \Delta Q_{изл} = \sigma S (T - T_0)^4 \Delta t, $$
где 
$S$ -- площадь поверхности электрического нагревателя,
$k$ -- коэффициент конвективного теплообмена, 
$T_0$ -- температура окружающей среды в кельвинах,
$ \sigma $ - постоянная Стефана-Больцмана, равная $5.67036713 \cdot 10^{-8} \frac{\textit{Вт}}{\textit{м}^2 \cdot \textit{K}^4}$.

Переобозначим ненадолго для удобства $Q_к + Q_{изл} = R(T) \Delta t$.

Мы нашли нашли формулу вычисления количества теплоты электрического нагревателя, выделенное за время $\Delta t$:
$$ Q = Q_н - R(T) \Delta t. $$ 

Зная её, можно легко вывести изменение температуры $\Delta T$ за $\Delta t$:
$$ cm \Delta T = Q = P \Delta t - R(T) \Delta t, $$
$$ \frac{\Delta T}{\Delta t} = \frac{P - R(T)}{cm}. $$
Устремляя $\Delta t \rightarrow 0 $ и добавляя начальное условие, получаем конечную модель.

**Математическая модель электрического нагревателя**:
$$
    \begin{cases}
        \frac{dT}{dt} = \frac{P - kS(T - T_0) - \sigma S (T^4 - T_0^4)}{cm}, \\
        T(0) = T_0.
    \end{cases}
$$

Добавим в электрический нагреватель реле температуры: если температура тела превышает определённое значение $T_{up}$, то нагрев приостанавливается, пока температура не упадёт до значения $T_d$. 

Введём такую функцию, которая будет описывать такое поведение:
$$
    H(T) = \begin{cases}
        0, T > T_{up}, \\
        1, T \leq T_d ~ и ~ \bar{H}(T) = 0,
    \end{cases}
$$
где $\bar{H}(T)$ - значение $H(T)$, вычисляемый за шаг до этого.

**Математическая модель электрического нагревателя с реле**:
$$
    \begin{cases}
        \frac{dT}{dt} = \frac{H(T) \cdot P - kS(T - T_0) - \sigma S (T^4 - T_0^4)}{cm}, \\
        T(0) = T_0.
    \end{cases}
$$

## Численное решение
Математическая модель представляет собой задачу Коши для обыкновенного дифференциального уравнения 1-го порядка, разрешенное относительно производной. Для решения можно использовать метод Рунге-Кутты четвертого порядка, который имеет погрешность порядка $O(h^4)$.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Константы 
SIGMA = 5.67036713e-8
K_CONST = 273.5
h = 0.1
```

```python
k = 10
p = 800
s = 0.5
T0 = 298
c = 1460
m = 1
```

```python
def f(t, T, a = 1):
    return (a * p - k*s*(T - T0) - SIGMA*s*(T**4 - T0**4)) /(c * m)
```

```python
def findFunction(x, y0, h):
    y = [y0]
    for i in range(len(x) - 1):
        k0 = f(x[i], y[i])
        k1 = f(x[i] + h/2, y[i] + h/2*k0)
        k2 = f(x[i] + h/2, y[i] + h/2*k1)
        k3 = f(x[i] + h, y[i] + h*k2)
        y.append(y[i] + h/6 * (k0 + 2*k1 + 2*k2 + k3))
    return y
```

```python
def findFunctionWithRelle(x, y0, h, down_bound, up_bound):
    y = [y0]
    a = 1
    for i in range(len(x) - 1):
        if y[i] > up_bound + K_CONST:
            a = 0
        elif y[i] < down_bound + K_CONST and a == 0:
            a = 1
        k0 = f(x[i], y[i], a)
        k1 = f(x[i] + h/2, y[i] + h/2*k0, a)
        k2 = f(x[i] + h/2, y[i] + h/2*k1, a)
        k3 = f(x[i] + h, y[i] + h*k2, a)
        y.append(y[i] + h/6 * (k0 + 2*k1 + 2*k2 + k3))
    return y
```

```python
t = np.array([t for t in np.arange(0, 4000, h)])
```

```python
T = findFunction(t, T0, h)

plt.plot(t, T - np.array(K_CONST))
plt.grid()
plt.xlabel('Время, t')
plt.ylabel('Температура, T')
plt.title('Зависимость температуры нагревателя от времени');
```

```python
up_bound = 175
down_bound = 120
T_relle = findFunctionWithRelle(t, T0, h, down_bound, up_bound)

plt.figure(figsize=(15, 7))
plt.plot(t, T_relle - np.array(K_CONST))
plt.grid()
plt.xlabel('Время, t')
plt.ylabel('Температура, T')
plt.title('Зависимость температуры нагревателя с релле от времени');
```

```python
m = 2
T1 = findFunction(t, T0, h)
m = 4
T2 = findFunction(t, T0, h)
```

```python
plt.plot(t, T - np.array(K_CONST))
plt.plot(t, T1 - np.array(K_CONST))
plt.plot(t, T2 - np.array(K_CONST))
plt.grid()
plt.legend(['m = 1', 'm = 2', 'm = 4'], title='Массы')
plt.xlabel('Время, t')
plt.ylabel('Температура, T')
plt.title('Сравнение в зависимости от масс');
```
